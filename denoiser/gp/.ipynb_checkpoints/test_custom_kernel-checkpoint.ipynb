```python
import numpy as np
from sklearn.gaussian_process.kernels import Kernel, RBF, WhiteKernel, Hyperparameter, ExpSineSquared
from scipy.spatial.distance import cdist
from sklearn.gaussian_process import GaussianProcessRegressor
from matplotlib import pyplot as plt
```

```python
from sklearn.gaussian_process.kernels import Kernel, Hyperparameter, RBF, WhiteKernel
from scipy.spatial.distance import cdist
import numpy as np

class CosineExpSquaredWhiteKernel(Kernel):
    def __init__(self, amplitude=1.0, length_scale=1.0, period=1.0, noise_level=1e-10,
                 amplitude_bounds=(1e-5, 1e5), length_scale_bounds=(1e-5, 1e5),
                 period_bounds=(1e-5, 1e5), noise_level_bounds=(1e-10, 1e1)):
        self.amplitude = amplitude
        self.length_scale = length_scale
        self.period = period
        self.noise_level = noise_level
        self.amplitude_bounds = amplitude_bounds
        self.length_scale_bounds = length_scale_bounds
        self.period_bounds = period_bounds
        self.noise_level_bounds = noise_level_bounds

        self.exp_squared_kernel = RBF(length_scale=length_scale)
        self.white_kernel = WhiteKernel(noise_level=noise_level)

    def __call__(self, X, Y=None, eval_gradient=False):
        X = np.atleast_2d(X)
        if Y is None:
            Y = X
        else:
            Y = np.atleast_2d(Y)

        dist = cdist(X, Y, metric='euclidean')
        cosine_component = np.cos(2 * np.pi * dist / self.period)
        exp_squared_component = self.exp_squared_kernel(X, Y)
        K = self.amplitude * cosine_component * exp_squared_component + self.white_kernel(X, Y)

        if eval_gradient:
            grad_amplitude = cosine_component * exp_squared_component
            grad_length_scale = self.amplitude * cosine_component * exp_squared_component * (dist ** 2) / (self.length_scale ** 3)
            grad_period = self.amplitude * exp_squared_component * (2 * np.pi * dist / self.period**2) * np.sin(2 * np.pi * dist / self.period)
            grad_noise_level = np.zeros_like(K)

            gradient = np.dstack([grad_amplitude, grad_length_scale, grad_period, grad_noise_level])
            return K, gradient

        return K

    def diag(self, X):
        return np.full(X.shape[0], self.amplitude) + self.noise_level

    def is_stationary(self):
        return False

    @property
    def hyperparameter_amplitude(self):
        return Hyperparameter("amplitude", "numeric", self.amplitude_bounds)

    @property
    def hyperparameter_length_scale(self):
        return Hyperparameter("length_scale", "numeric", self.length_scale_bounds)

    @property
    def hyperparameter_period(self):
        return Hyperparameter("period", "numeric", self.period_bounds)

    @property
    def hyperparameter_noise_level(self):
        return Hyperparameter("noise_level", "numeric", self.noise_level_bounds)

    def get_params(self, deep=True):
        return {
            "amplitude": self.amplitude,
            "length_scale": self.length_scale,
            "period": self.period,
            "noise_level": self.noise_level,
            "amplitude_bounds": self.amplitude_bounds,
            "length_scale_bounds": self.length_scale_bounds,
            "period_bounds": self.period_bounds,
            "noise_level_bounds": self.noise_level_bounds,
        }

    def set_params(self, **params):
        for param, value in params.items():
            setattr(self, param, value)
        return self
```

```python
class QuasiperiodicKernel(Kernel):
    def __init__(self, amplitude=1.0, length_scale=1.0, period=1.0, omega=1.0,  noise_level=1e-10,
                 amplitude_bounds=(1e-5, 1e5), length_scale_bounds=(1e-5, 1e5),
                 period_bounds=(1e-5, 1e5), noise_level_bounds=(1e-10, 1e1)):
        self.amplitude = amplitude
        self.length_scale = length_scale
        self.period = period
        self.omega = omega
        self.noise_level = noise_level
        
        self.amplitude_bounds = amplitude_bounds
        self.length_scale_bounds = length_scale_bounds
        self.period_bounds = period_bounds
        self.noise_level_bounds = noise_level_bounds
        
        self.exp_sine_squared_kernel = ExpSineSquared(length_scale=length_scale, periodicity=period)
        self.white_kernel = WhiteKernel(noise_level=noise_level)

    def __call__(self, X, Y=None, eval_gradient=False):
        X = np.atleast_2d(X)
        if Y is None:
            Y = X
        else:
            Y = np.atleast_2d(Y)

        dist = cdist(np.ceil(X / period), np.ceil(Y / period) , metric='minkowski', p=1.0)
        omega_component = omega ** dist
        local_periodic_kernel = self.exp_sine_squared_kernel(X, Y)
        white_kernel = self.white_kernel(X, Y)
        K = (self.amplitude * omega_component * local_periodic_kernel) + white_kernel

        if eval_gradient:
            grad_amplitude = omega_component * local_periodic_kernel + white_kernel
            grad_length_scale = K * 4 * np.sin(np.pi * cdist(X, Y, metric='euclidean') /  period)**2 / length_scale**3 #add in omega term
            grad_period = K * (-4 * np.sin(np.pi * cdist(X, Y, metric='euclidean') / period) * np.cos(np.pi * cdist(X, Y, metric='euclidean') / period) * (-np.pi * cdist(X, Y, metric='euclidean') / period**2 ))
            grad_omega = self.amplitude * (dist * omega ** (dist - 1)) * local_periodic_kernel + white_kernel
            grad_noise_level = np.zeros_like(K)

            gradient = np.dstack([grad_amplitude, grad_length_scale, grad_period, grad_noise_level])
            return K, gradient

        return K

    def diag(self, X):
        return np.full(X.shape[0], self.amplitude) + self.noise_level

    def is_stationary(self):
        return False

    @property
    def hyperparameter_amplitude(self):
        return Hyperparameter("amplitude", "numeric", self.amplitude_bounds)

    @property
    def hyperparameter_length_scale(self):
        return Hyperparameter("length_scale", "numeric", self.length_scale_bounds)

    @property
    def hyperparameter_period(self):
        return Hyperparameter("period", "numeric", self.period_bounds)

    @property
    def hyperparameter_noise_level(self):
        return Hyperparameter("noise_level", "numeric", self.noise_level_bounds)

    def get_params(self, deep=True):
        return {
            "amplitude": self.amplitude,
            "length_scale": self.length_scale,
            "period": self.period,
            "noise_level": self.noise_level,
            "amplitude_bounds": self.amplitude_bounds,
            "length_scale_bounds": self.length_scale_bounds,
            "period_bounds": self.period_bounds,
            "noise_level_bounds": self.noise_level_bounds,
        }

    def set_params(self, **params):
        for param, value in params.items():
            setattr(self, param, value)
        return self
```

```python
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(X).ravel() + 0.1 * np.random.normal(size=X.shape[0]) 
```

```python
# Instantiate the custom kernel
kernel = CosineExpSquaredWhiteKernel(
    amplitude=1.0, length_scale=1.0, period=1.0, noise_level=np.std(y),
    amplitude_bounds=(1e-3, 1e3), length_scale_bounds=(1e-3, 1e3),
    period_bounds=(1e-3, 1e3), noise_level_bounds=(1e-3, 1e1)
)
```

```python
# Create the Gaussian Process model
gp = GaussianProcessRegressor(kernel=kernel)
```

```python
# Fit the model to the data
gp.fit(X, y)
```

```python
# Make predictions
X_pred = np.linspace(0, 10, 100).reshape(-1, 1)
y_pred, sigma = gp.predict(X_pred, return_std=True)
```

```python
# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(X, y, 'r.', markersize=5, label='Noisy observations')
plt.plot(X, np.sin(X), 'g*', markersize=5, label='True observations')
plt.plot(X_pred, y_pred, 'b-', label='Prediction')
plt.fill_between(X_pred.ravel(), y_pred - 1.96 * sigma, y_pred + 1.96 * sigma, alpha=0.2, color='k')
plt.title('Gaussian Process with Cosine * Exponential Squared + White Kernel')
plt.legend()
```
