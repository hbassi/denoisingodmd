```python
import numpy as np
from scipy.linalg import svd, hankel, eig
from matplotlib import pyplot as plt
import scipy
from tqdm import trange
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C, ExpSineSquared, WhiteKernel, Kernel, Hyperparameter
import jax.numpy as jnp
from scipy.spatial.distance import cdist
from scipy.io import savemat, loadmat
```

```python
np.random.seed(999)
```

```python
noise = 0.1
overlap = 0.2
Tmax = 1000
dt = 1
```

```python
def generate_samples(E, psi0, dt=1, nb=100):
    S = np.zeros(nb, dtype=np.complex128)
    
    for j in trange(nb):
        S[j] = np.sum(np.abs(psi0)**2 * np.exp(-1j * E * j * dt))
    
    return S
```

```python
def generate_samples_der(E, psi0, dt=1, nb=100, n=1):
    S = np.zeros(nb, dtype=np.complex128)
    
    for j in trange(nb): 
        S[j] = np.sum(np.abs(psi0)**2 * (-1j * E)**n * np.exp(-1j * E * j * dt))
    
    return S
```

```python
def lam2lamt(lam, lammin, lammax):
    lamt = np.pi / 2 * (lam - (lammin + lammax) / 2) / (lammax - lammin)
    return lamt
```

```python
def lamt2lam(lamt, lammin, lammax):
    lam = lamt * 2 / np.pi * (lammax - lammin) + (lammin + lammax) / 2
    return lam
```

```python
def generate_phi(overlap, N):
    phi = np.zeros((N,1))
    phi[0] = np.sqrt(overlap)
    phi[1:] = np.sqrt((1 - phi[0]**2) / (N - 1))
    # phi[1] = np.sqrt(overlap)
    # phi[0] = np.sqrt((1 - phi[1]**2) / (N - 1))
    # phi[2:] = np.sqrt((1 - phi[1]**2) / (N - 1))
    return phi
```

```python
data_cr2 = scipy.io.loadmat('../../data/Cr2_4000.mat')
psiHF = data_cr2['psiHF']
E = data_cr2['E']
```

```python
Et = lam2lamt(E,E[0],E[-1])
# E_center = E - (E.min() + E.max())/2
# Et = E_center / np.abs(E_center).max()
```

```python
phi = generate_phi(overlap,len(Et))
print('generated phi')
dataS = generate_samples_der(Et,phi,dt,Tmax, n=0)
```

```python
plt.plot(dataS.real)
plt.xlabel('t')
plt.ylabel('o(t)')
plt.title('True dynamics')
```

```python
tdataS = dataS + noise * np.random.randn(Tmax) + 1j * noise * np.random.randn(Tmax)
tdataS = np.clip(tdataS, -1, 1)
```

```python
tdataS[0] = 1.0 + 0.0j
plt.plot(tdataS.imag)
tdataS.shape
#plt.plot(tdataS.real)
```

```python
noisydataS = np.array(tdataS.real)
noisydataS.shape
```

```python
t = np.arange(0, Tmax, 1) * dt
```

```python
# Fit GP to the signals
y = noisydataS.real
#X = np.tile(t, (len(y), 1))  # Repeat t to create as many rows as in y
X = t[:, np.newaxis]
#y = average_noisy_signal
X.shape, y.shape
```

```python
class CosineExpSquaredWhiteKernel(Kernel):
    def __init__(self, amplitude=1.0, length_scale=1.0, period=1.0, noise_level=0.1,
                 amplitude_bounds=(1e-5, 1e5), length_scale_bounds=(1e-5, 1e5),
                 period_bounds=(1e-5, 1e5), noise_level_bounds=(1e-10, 1e1)):
        self.amplitude = amplitude
        self.length_scale = length_scale
        self.period = period
        self.noise_level = noise_level
        self.amplitude_bounds = amplitude_bounds
        self.length_scale_bounds = length_scale_bounds
        self.period_bounds = period_bounds
        self.noise_level_bounds = noise_level_bounds

        self.exp_squared_kernel = RBF(length_scale=length_scale)
        self.white_kernel = WhiteKernel(noise_level=noise_level)

    def __call__(self, X, Y=None, eval_gradient=False):
        X = np.atleast_2d(X)
        if Y is None:
            Y = X
        else:
            Y = np.atleast_2d(Y)

        dist = cdist(X, Y, metric='euclidean')
        cosine_component = np.cos(2 * np.pi * dist / self.period)
        exp_squared_component = self.exp_squared_kernel(X, Y)
        K = self.amplitude * cosine_component * exp_squared_component + self.white_kernel(X, Y)

        if eval_gradient:
            grad_amplitude = cosine_component * exp_squared_component
            grad_length_scale = self.amplitude * cosine_component * exp_squared_component * (dist ** 2) / (self.length_scale ** 3)
            grad_period = self.amplitude * exp_squared_component * (2 * np.pi * dist / self.period**2) * np.sin(2 * np.pi * dist / self.period)
            grad_noise_level = np.zeros_like(K)

            gradient = np.dstack([grad_amplitude, grad_length_scale, grad_period, grad_noise_level])
            return K, gradient

        return K

    def diag(self, X):
        return np.full(X.shape[0], self.amplitude) + self.noise_level

    def is_stationary(self):
        return False

    @property
    def hyperparameter_amplitude(self):
        return Hyperparameter("amplitude", "numeric", self.amplitude_bounds)

    @property
    def hyperparameter_length_scale(self):
        return Hyperparameter("length_scale", "numeric", self.length_scale_bounds)

    @property
    def hyperparameter_period(self):
        return Hyperparameter("period", "numeric", self.period_bounds)

    @property
    def hyperparameter_noise_level(self):
        return Hyperparameter("noise_level", "numeric", self.noise_level_bounds)

    def get_params(self, deep=True):
        return {
            "amplitude": self.amplitude,
            "length_scale": self.length_scale,
            "period": self.period,
            "noise_level": self.noise_level,
            "amplitude_bounds": self.amplitude_bounds,
            "length_scale_bounds": self.length_scale_bounds,
            "period_bounds": self.period_bounds,
            "noise_level_bounds": self.noise_level_bounds,
        }

    def set_params(self, **params):
        for param, value in params.items():
            setattr(self, param, value)
        return self
```

```python
kernel = CosineExpSquaredWhiteKernel(
    amplitude=1.0, length_scale=1.0, period=1.0, noise_level=np.std(y),
    amplitude_bounds=(1e-3, 1e3), length_scale_bounds=(1e-3, 1e2),
    period_bounds=(1e-3, 1e3), noise_level_bounds=(1e-3, 1e0)
)
```

```python
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=50, alpha=1e-5)
gp.fit(X, y)
```

```python
# Predict
mean_prediction, std_prediction = gp.predict(X, return_std=True)
```
