```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a true sine wave
def generate_sine_wave(t):
    return np.sin(t)

# Function to generate noisy observations of the sine wave
def generate_noisy_sine_wave(true_wave, noise_std):
    return true_wave + np.random.normal(0, noise_std, size=true_wave.shape)

# Kalman Filter implementation
def kalman_filter(noisy_data, process_variance, measurement_variance):
    n = len(noisy_data)
    estimated_state = np.zeros(n)
    estimated_error = np.zeros(n)
    
    # Initial guesses
    estimated_state[0] = noisy_data[0]
    estimated_error[0] = 1.0
    
    for k in range(1, n):
        # Time update (predict)
        predicted_state = estimated_state[k-1]
        predicted_error = estimated_error[k-1] + process_variance
        
        # Measurement update (correct)
        kalman_gain = predicted_error / (predicted_error + measurement_variance)
        estimated_state[k] = predicted_state + kalman_gain * (noisy_data[k] - predicted_state)
        estimated_error[k] = (1 - kalman_gain) * predicted_error
    
    return estimated_state

# Generate the time axis
t = np.linspace(0, 10, 1000)
true_sine_wave = generate_sine_wave(t)

# Parameters
num_trajectories = 2000
noise_std = 0.5
process_variance = 1e-4
measurement_variance = noise_std**2

# Generate multiple noisy sine waves and apply Kalman filter
all_noisy_sine_waves = np.zeros((num_trajectories, len(t)))
all_estimated_sine_waves = np.zeros((num_trajectories, len(t)))

for i in range(num_trajectories):
    noisy_sine_wave = generate_noisy_sine_wave(true_sine_wave, noise_std)
    estimated_sine_wave = kalman_filter(noisy_sine_wave, process_variance, measurement_variance)
    all_noisy_sine_waves[i, :] = noisy_sine_wave
    all_estimated_sine_waves[i, :] = estimated_sine_wave

# Compute the average estimated sine wave
average_estimated_sine_wave = np.mean(all_estimated_sine_waves, axis=0)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(t, true_sine_wave, label='True Sine Wave', linestyle='--')
plt.plot(t, np.mean(all_noisy_sine_waves, axis=0), label='Average Noisy Sine Wave', alpha=0.5)
plt.plot(t, average_estimated_sine_wave, label='Average Estimated Sine Wave (Kalman Filter)', linewidth=2)
plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Denoising Multiple Noisy Sine Waves using Kalman Filter')
plt.show()
```
