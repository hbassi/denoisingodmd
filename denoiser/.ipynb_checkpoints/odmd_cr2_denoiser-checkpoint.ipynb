```python
import numpy as np
from scipy.linalg import svd, hankel, eig
from matplotlib import pyplot as plt
import scipy
from tqdm import trange
import torch
import torch.nn as nn
import torch.nn.init as init
```

```python
def dmd(data, dt, tol=1e-6):
    
    k = len(data)
    X = make_hankel(data, int(np.floor(k / 3) + 1), int(np.ceil(2 / 3 * k)))

    X1 = X[:, :-1]
    X2 = X[:, 1:]

    U, S, Vh = svd(X1, full_matrices=False)
    V = Vh.conj().T

    r = np.sum(S > tol * S[0])
    U = U[:, :r]
    S = np.diag(S[:r])
    V = V[:, :r]

    Atilde = U.conj().T @ X2 @ V @ np.linalg.inv(S)
    mu = eig(Atilde)[0]
    omega = np.log(mu) / dt

    return omega
```

```python
def generate_samples(E, psi0, dt=1, nb=100):
    H = np.zeros(nb, dtype=np.complex128)
    S = np.zeros(nb, dtype=np.complex128)
    
    for j in trange(nb):
        H[j] = np.sum(E * np.abs(psi0)**2 * np.exp(-1j * E * j * dt))
        S[j] = np.sum(np.abs(psi0)**2 * np.exp(-1j * E * j * dt))
    
    return H, S
```

```python
def lam2lamt(lam, lammin, lammax):
    lamt = np.pi / 2 * (lam - (lammin + lammax) / 2) / (lammax - lammin)
    return lamt
```

```python
def lamt2lam(lamt, lammin, lammax):
    lam = lamt * 2 / np.pi * (lammax - lammin) + (lammin + lammax) / 2
    return lam
```

```python
def make_hankel(data, m, n):
    return hankel(data[:m], data[m-1:m+n-1])
```

```python
def generate_phi(overlap, N):
    phi = np.zeros((N,1))
    phi[0] = np.sqrt(overlap)
    phi[1:] = np.sqrt((1 - phi[0]**2) / (N - 1))
    return phi
```

```python
def run_compare(dataH, dataS, dt, fun=dmd, tol=[1e-1, 1e-2, 1e-3], Tmax=500, step=1):
    t = np.arange(10, Tmax + 1, step)
    lam = np.inf * np.ones((len(t), len(tol)))
    
    for i in range(len(t)):
        for j in range(len(tol)):
            omega = fun(dataS[:t[i]], dt, tol[j])
            lam[i, j] = -np.max(np.imag(omega))
            
    return lam, t
```

```python
#load the model
torch.manual_seed(999)
class DenoisingAutoencoder(nn.Module):
    def __init__(self, input_size, hidden_size, latent_size):
        super(DenoisingAutoencoder, self).__init__()
        
        self.encoder = nn.Sequential(
            nn.Linear(input_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, latent_size).double()
        )

    
        self.decoder = nn.Sequential(
            nn.Linear(latent_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, hidden_size).double(),
            nn.Tanh(),
            nn.Linear(hidden_size, input_size).double()
        )


    def forward(self, x):
        latent = self.encoder(x)
        reconstruction = self.decoder(latent)
        return reconstruction
```

```python
model = DenoisingAutoencoder(input_size, hidden_size, latent_size).to(device)
model.load_state_dict(torch.load('denoiser_numtraj=2000_noise=1.0.pth'))
model.eval()
```

```python
denoised_dataS = model(torch.tensor(noisydataS).double().to(device))
```

```python
denoised_dataSp = denoised_dataS.cpu().detach().numpy()
```

```python
lamt,t = run_compare(dataH, denoised_dataSp, dt,fun, tol,Tmax,step)
```

```python
lam = lamt2lam(lamt,E[0],E[-1])
```

```python
def plot_compare(t, lam, tol, E, mytitle='', xlimits=None, ylimits=None):
    plt.figure()
    marker = '*ods'
    lgnd = []
    for i in range(len(tol)):
        mark = marker[i % len(marker)]
        plt.semilogy(t, np.abs(lam[:, i] - E[0]), mark, label=f'tol = {tol[i]}')
    
    plt.plot([0, t[-1]], [1e-3, 1e-3], ':k')
    plt.legend()
    plt.xlabel('# timesteps')
    plt.ylabel('absolute error')
    if mytitle:
        plt.title(mytitle)
    if xlimits:
        plt.xlim(xlimits)
    if ylimits:
        plt.ylim(ylimits)
    plt.show()
```

```python
mytitle = molecule + ' (noise = '+ str(noise) + ')'
plot_compare(t,lam, tol ,E,mytitle,[0,Tmax],[1e-6,10])
```
