```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
# Noise levels and threshold
noises = [0.8, 0.1, 0.01]
threshold = 1e-3

# SVD tolerance values
svd_tolerances = [0.8, 0.1, 0.01, 0.001]

# Number of rows in the data arrays (tolerances)
num_tolerances = len(svd_tolerances)

# Initialize storage for indices
indices_baseline = np.full((num_tolerances, len(noises)), -1, dtype=int)  # -1 if no value is below threshold
indices_fd = np.full((num_tolerances, len(noises)), -1, dtype=int)

for j, noise in enumerate(noises):
    # Load data
    baseline = np.load(f'baseline_odmd_noise={noise}_Tmax=1000_overlap=0.2_dt=1.0_real_ODMD.npy')
    fd = np.load(f'fourier_denoised_odmd_3mode_noise={noise}_Tmax=1000_overlap=0.2_dt=1.0_real_stacked_reduced=0.01_test_ODMD.npy')

    # Find indices below the threshold for each tolerance row
    for i in range(num_tolerances):
        below_threshold_baseline = np.where(baseline[i, :] < threshold)[0]
        below_threshold_fd = np.where(fd[i, :] < threshold)[0]

        # Record the first index below the threshold if it exists
        if len(below_threshold_baseline) > 0:
            indices_baseline[i, j] = below_threshold_baseline[0]
        if len(below_threshold_fd) > 0:
            indices_fd[i, j] = below_threshold_fd[0]

# Helper function to mask invalid tiles
def apply_mask(data):
    masked_data = np.ma.masked_where(data == -1, data)
    return masked_data

# Apply mask to the data
masked_baseline = apply_mask(indices_baseline)
masked_fd = apply_mask(indices_fd)

# Plotting
fig = plt.figure(figsize=(14, 7))
gs = GridSpec(1, 3, width_ratios=[1, 1, 0.05], wspace=0.3)

# Heatmap for baseline
ax1 = fig.add_subplot(gs[0, 0])
im1 = ax1.imshow(masked_baseline * 10, cmap='viridis', aspect='auto')
ax1.set_title('Baseline Heatmap')
ax1.set_xlabel('Noise Levels')
ax1.set_ylabel('SVD Tolerance')
ax1.set_xticks(range(len(noises)))
ax1.set_xticklabels([f'{n:.2f}' for n in noises])
ax1.set_yticks(range(num_tolerances))
ax1.set_yticklabels([f'{t:.3f}' for t in svd_tolerances])

# Annotate each tile with the index (multiplied by 10)
for i in range(num_tolerances):
    for j in range(len(noises)):
        value = indices_baseline[i, j]
        if value != -1:  # Only annotate valid tiles
            ax1.text(j, i, f'{value * 10}', ha='center', va='center', color='white')

# Heatmap for Fourier denoised
ax2 = fig.add_subplot(gs[0, 1])
im2 = ax2.imshow(masked_fd * 10, cmap='viridis', aspect='auto')
ax2.set_title('Fourier Denoised Heatmap')
ax2.set_xlabel('Noise Levels')
ax2.set_ylabel('SVD Tolerance')
ax2.set_xticks(range(len(noises)))
ax2.set_xticklabels([f'{n:.2f}' for n in noises])
ax2.set_yticks(range(num_tolerances))
ax2.set_yticklabels([f'{t:.3f}' for t in svd_tolerances])

# Annotate each tile with the index (multiplied by 10)
for i in range(num_tolerances):
    for j in range(len(noises)):
        value = indices_fd[i, j]
        if value != -1:  # Only annotate valid tiles
            ax2.text(j, i, f'{value * 10}', ha='center', va='center', color='white')

# Add a single colorbar for both plots
cbar_ax = fig.add_subplot(gs[0, 2])
cbar = fig.colorbar(im2, cax=cbar_ax)
cbar.set_label('# Observables')


plt.tight_layout()
#plt.savefig('baseline_vs_FD_heatmaps'+'_Tmax=1000_overlap=0.2_dt=1.0_real_ODMD.png')
plt.show()
```
