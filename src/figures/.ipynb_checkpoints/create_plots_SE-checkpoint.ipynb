```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
numpads = [0, 4, 16, 64]
noise = 0.8
t = np.arange(0, 1500, 1)

# Define a list of unique markers
markers = ['o', 's', 'D', 'v', 'x', '<', '>', 'P']
# Extend the colormap to accommodate all plots (numpads + baseline + extra dataset)
colors = cm.tab10(np.linspace(0, 1, len(numpads) + 3))


fig, ax = plt.subplots(figsize=(10, 6))

# Plot data for each numpad
for i, numpad in enumerate(numpads):
    fd_filename = f'FFT_zero_padding_noise={noise}_Tmax=2000_overlap=0.2_dt=1.0_padding={numpad}_ff_left_right_absolute_errors_real_dt=1.0test1.npy'
    data = np.load(fd_filename, allow_pickle=True)
    data = data[:1500]
    ax.semilogy(
        t[::3],
        data[::3],
        marker=markers[i % len(markers)],
        color=colors[i],
        linestyle='',
        label=fr'$M$ = {numpad}'
    )

# Plot baseline ODMD for noise = 0.8
baseline_filename = f'baseline_odmd_noise={noise}_Tmax=1500_overlap=0.2_dt=1.0_real_ODMD.npy'
idx = 0
t1 = np.arange(20, 1500, 1)
baseline_data = np.load(baseline_filename)
baseline_data = baseline_data[idx][:1500]
ax.semilogy(
    t1[::3],
    baseline_data[::3],
    marker='X',
    color='black',
    linestyle='',
    label='Baseline ODMD'
)

# Plot FDODMD for noise = 0.8
extra_filename = 'Cr2_fourier_denoised_odmd_6_stacked_gamma=3.0_denoised_noise=0.8_Tmax=1500_overlap=0.2_dt=1.0_tol=0.8-0.1_real_ODMD.npy'
extra_data = np.load(extra_filename)
t2 = np.arange(25, 1500, 5)
extra_data = extra_data[-1][:1500]
ax.semilogy(
    t2[::],
    extra_data[:295],
    marker=markers[5],
    color=colors[len(numpads) + 1],
    linestyle='',
    label=r'FDODMD w/o $d(t)$'
)
denoised_with_noise = 'Cr2_fourier_denoised_odmd_6_stacked_denoised_noise=0.8_Tmax=1500_overlap=0.2_dt=1.0_tol=0.8-0.8_real_ODMD.npy'
denoised_w_noise_data = np.load(denoised_with_noise)
denoised_w_noise_data = denoised_w_noise_data[-1][:1500]
ax.semilogy(
    t2[::],
    denoised_w_noise_data[:295],
    marker=markers[6],
    color=colors[len(numpads) + 2],
    linestyle='',
    label=r'FDODMD w/ $d(t)$'
)
# Plot chemical accuracy line
ax.plot(
    [0, 1499],
    [1e-3, 1e-3],
    'k:'
)

# Set plot attributes with styled fonts and ticks
ax.set_ylim([1e-7, 1e-1])
ax.set_xlabel('t', fontsize=18, fontweight='bold')
ax.set_ylabel('absolute error', fontsize=18, fontweight='bold')
ax.set_title(f'Zero-padding vs FDODMD vs ODMD', fontsize=18, fontweight='bold', pad=10)
ax.tick_params(axis='both', which='major', labelsize=13)
ax.legend(loc='lower left', frameon=False, fontsize=14)

plt.tight_layout()
#plt.savefig('Cr2_fourier_denoised_odmd_vs_odmd_vs_zeropadding_noise=0.8_styled.pdf', bbox_inches='tight')
plt.show()
```

```python
# Define a list of unique markers and a colormap
markers = ['o', 's', 'D', 'v', '^']
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
numpad = 64
noise_levels = [1.0, 0.8, 0.5, 0.1, 0.0]
t = np.arange(0, 1500, 1)

fig, ax = plt.subplots(figsize=(10, 6))

# Plot data for each noise level
for i, noise in enumerate(noise_levels):
    fd_filename = f'FFT_zero_padding_noise={noise}_Tmax=2000_overlap=0.2_dt=1.0_padding={numpad}_ff_left_right_absolute_errors_real_dt=1.0test1.npy'
    data = np.load(fd_filename, allow_pickle=True)
    data = data[:1500]
    ax.semilogy(
        t,
        data,
        marker=markers[i % len(markers)],
        color=colors[i],
        linestyle='',
        label=fr'$\epsilon$ = {noise}'
    )

# Plot chemical accuracy line
ax.plot(
    [0, 1499],
    [1e-3, 1e-3],
    'k:'
)

ax.set_ylim([1e-7, 1e-1])
ax.set_xlabel('t', fontsize=18, fontweight='bold')
ax.set_ylabel('absolute error', fontsize=18, fontweight='bold')
ax.set_title(fr'Zero-padding, $M$ = {numpad}', fontsize=18, fontweight='bold', pad=10)
ax.tick_params(axis='both', which='major', labelsize=13)
ax.legend(loc='upper right', frameon=False, fontsize=12)

plt.tight_layout()
#plt.savefig(f'FFT_zero_padding_vs_odmd_baseline_numpad_{numpad}_styled.pdf', bbox_inches='tight')
plt.show()
```
