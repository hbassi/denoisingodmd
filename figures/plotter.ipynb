```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import loadmat
```

```python
pads = [0, 1, 2, 3, 4, 5]
```

```python
baseline_odmd_best = loadmat('Cr2-n0.1-o20-tol0.8_________0.1________0.01_______0.001-Tmax1000-dt1-_vanilla_complex.mat')['odmd_errors']
```

```python
for pad in pads:
    data = np.load('FFT_cr2_gse_errors_numpad='+str(pad)+'_noise=0.1_multiple_pad_ff_absolute_errors.npy')
    plt.semilogy(np.arange(15, 1000), data, '*', label='Numpad = '+ str(pad))
plt.semilogy(np.arange(15, 1001), baseline_odmd_best, 'o', label='ODMD')
plt.plot([15, 1000], [1e-3, 1e-3], 'k:', label='chemical accuracy')
plt.legend(frameon=False)
plt.xlabel('# observables')
plt.ylabel('absolute error')
plt.ylim([1e-7, 1e-1]);
```
