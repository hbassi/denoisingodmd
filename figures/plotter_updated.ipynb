```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
pads = [0, 1, 2, 3, 4, 5]
for pad in pads:
    data = np.load(f'FFT_zero_padding_noise=0.1_Tmax=1000_overlap=0.2_dt=1_padding={pad}_ff_left_right_absolute_errors_real_test1.npy')
    plt.semilogy(np.arange(20, 1001), data , '*', label='Numpad = '+ str(pad))
baseline = np.load('baseline_odmd_noise=0.1_Tmax=1000_overlap=0.2_dt=1_real_ODMD.npy')
plt.semilogy(np.arange(20, 1001), baseline[0, :], 'o', label='ODMD')
plt.plot([20, 1000], [1e-3, 1e-3], 'k:', label='chemical accuracy')
plt.legend(frameon=False)
plt.xlabel('# observables')
plt.ylabel('absolute error')
plt.ylim([1e-7, 1e-1]);
```
