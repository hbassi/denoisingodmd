```python
import numpy as np
from scipy.linalg import svd, hankel, eig
from matplotlib import pyplot as plt
import scipy
from tqdm import trange
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class StackedLSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(StackedLSTMModel, self).__init__()
        self.hidden_size = hidden_size
        
        self.lstm1 = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.lstm2 = nn.LSTM(hidden_size, hidden_size, batch_first=True)
    
        self.fc = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        h0_1 = torch.zeros(1, x.size(0), self.hidden_size, dtype=torch.float32).to(x.device)
        c0_1 = torch.zeros(1, x.size(0), self.hidden_size, dtype=torch.float32).to(x.device)
        h0_2 = torch.zeros(1, x.size(0), self.hidden_size, dtype=torch.float32).to(x.device)
        c0_2 = torch.zeros(1, x.size(0), self.hidden_size, dtype=torch.float32).to(x.device)
    

        out, _ = self.lstm1(x, (h0_1, c0_1))
        out, _ = self.lstm2(out, (h0_2, c0_2))
        
        out = self.fc(out[:, -1, :])
        
        return out
```

```python
device = 'cpu'
input_size = 750
hidden_size = 1028
output_size = 1
num_epochs = 500
learning_rate = 0.01
```

```python
model = StackedLSTMModel(input_size, hidden_size, output_size)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
def generate_samples(E, psi0, dt=1, nb=100):
    H = np.zeros(nb, dtype=np.complex128)
    S = np.zeros(nb, dtype=np.complex128)
    
    for j in trange(nb):
        H[j] = np.sum(E * np.abs(psi0)**2 * np.exp(-1j * E * j * dt))
        S[j] = np.sum(np.abs(psi0)**2 * np.exp(-1j * E * j * dt))
    
    return H, S
```

```python
def lam2lamt(lam, lammin, lammax):
    lamt = np.pi / 2 * (lam - (lammin + lammax) / 2) / (lammax - lammin)
    return lamt
```

```python
def lamt2lam(lamt, lammin, lammax):
    lam = lamt * 2 / np.pi * (lammax - lammin) + (lammin + lammax) / 2
    return lam
```

```python
def make_hankel(data, m, n):
    return hankel(data[:m], data[m-1:m+n-1])
```

```python
def generate_phi(overlap, N):
    phi = np.zeros((N,1))
    phi[0] = np.sqrt(overlap)
    phi[1:] = np.sqrt((1 - phi[0]**2) / (N - 1))
    return phi
```

```python
seed = 100
np.random.seed(seed)
```

```python
data_cr2 = scipy.io.loadmat('./data/Cr2_4000.mat')
```

```python
psiHF = data_cr2['psiHF']
E = data_cr2['E']
```

```python
molecule = 'Cr2'
noise = 0.999
overlap = 0.2# 0: HF
tol = 0.1
Tmax = 750
step = 1
dt = 1
```

```python
Et = lam2lamt(E,E[0],E[-1])
```

```python
if overlap == 0:
    dataH,dataS = generate_samples(Et,psiHF,dt,Tmax)
else:
    phi = generate_phi(overlap,len(Et));
    print('generated phi')
    [dataH,dataS] = generate_samples(Et,phi,dt,Tmax);
```

```python
plt.plot(dataS.real)
plt.xlabel('t')
plt.ylabel('o(t)')
plt.title('True dynamics')
```

```python
dataH = (dataH + noise * np.random.randn(Tmax) + 1j * noise * np.random.randn(Tmax)).real
dataS = (dataS + noise * np.random.randn(Tmax) + 1j * noise * np.random.randn(Tmax)).real
```

```python
plt.plot(dataS.real)
plt.xlabel('t')
plt.ylabel('o(t)')
plt.title('Noisy dynamics')
```

```python
# (num_samples, sequence_length, input_size)
dataS = dataS.reshape((1,1,750))
```

```python
GSE = np.array(E[0]).reshape((1,1))
GSE.shape
```

```python
dataS = torch.tensor(dataS).to(torch.float32).to(device)
GSE = torch.tensor(GSE).to(torch.float32).to(device)
```

```python
for epoch in trange(num_epochs):
    model.train()
    
    outputs = model(dataS)
    loss = criterion(outputs, GSE)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.8f}')

print("Training complete")
```

```python
model.eval()
```

```python
clean_dataS = torch.tensor(generate_samples(Et,phi,dt,Tmax)[1]).real.reshape((1,1,dataS.shape[2]))
```

```python
np.random.seed(999)
errs = []
pred_eval = []
noisy_data = clean_dataS + 0.999 * np.random.randn(dataS.shape[2])
for k in trange(50,750):
    padded_new_noisy_data = -np.ones(dataS.shape)
    cutoff = k
    padded_new_noisy_data[:,:,:cutoff] = (noisy_data)[0,0,:cutoff]
    with torch.no_grad():  
        prediction = model(torch.tensor(padded_new_noisy_data).type(torch.float32))
    abserr = abs(prediction - E[0])
    pred_eval.append(prediction.item())
    errs.append(abserr.item())
```

```python
plt.scatter(list(range(50,750)),errs,marker='*')
plt.plot([50, 750], [1e-3, 1e-3], ':k')
plt.xlabel('Length of queried sequence')
plt.ylabel(r'$|\widetilde{GSE} - GSE|$')
plt.yscale('log')
plt.title('LSTM Prediction of GSE (noise = '+str(0.999)+ ', overlap = '+str(overlap) + ')')
plt.legend(['abs err','chemical accuracy'],frameon=True)
plt.tight_layout()
#plt.savefig('./figures/rnn_cr2_train_noise=0.999_overlap=0.2_test_clean_noise=0.999_seed=999.pdf')
```

```python
plt.scatter(list(range(50,750)),pred_eval,marker='*')
plt.plot([50, 750], [E[0], E[0]], ':k')
plt.xlabel('Length of queried sequence')
plt.ylabel(r'$\lambda_0$')
#plt.yscale('log')
plt.title('LSTM Prediction of GSE (noise = '+str(0.999)+ ', overlap = '+str(overlap) + ')')
plt.legend([r'predicted $\lambda_0$',r'ground truth $\lambda_0$'],frameon=True)
plt.tight_layout()
#plt.savefig('./figures/rnn_cr2_train_noise=0.999_overlap=0.2_test_clean_noise=0.999_seed=999.pdf')
```

```python
all_noises = [2, 4, 8, 10, 12, 16, 18, 20, 24]
fig, axes = plt.subplots(3, 3, figsize=(12, 12))



for i in range(len(all_noises)):
    high_noise_errs = []
    noise = all_noises[i]
    np.random.seed(999)
    noisy_data = clean_dataS + noise * np.random.randn(dataS.shape[2])
    
    for k in trange(50,750):
        padded_new_noisy_data = -np.ones(dataS.shape)
        cutoff = k
        padded_new_noisy_data[0,0,:cutoff] = noisy_data[0,0,:cutoff]
        
        with torch.no_grad():  
            prediction = model(torch.tensor(padded_new_noisy_data).type(torch.float32))
            
        abserr = abs(prediction - E[0])
        high_noise_errs.append(abserr.item())
        
    row, col = divmod(i, 3)
    axes[row, col].scatter(list(range(50, 750)), high_noise_errs, marker='*')
    axes[row, col].plot([50, 750], [1e-3, 1e-3], ':k')
    axes[row, col].set_xlabel('Length of queried sequence')
    axes[row, col].set_ylabel(r'$|\widetilde{GSE} - GSE|$')
    axes[row, col].set_yscale('log')
    axes[row, col].set_title(f'test noise = {noise}, overlap = 0.2')
    axes[row, col].legend(['abs err', 'chemical accuracy'], frameon=False)
    fig.tight_layout()
    #plt.savefig('./figures/rnn_cr2_train_noise=0.999_overlap=0.2_test_clean_noises_2-24_seed=999.pdf')
```
